Import pandas, matplotlib, and the date module for matplotlib.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
%matplotlib inline
plt.rc('font', serif='Arial') 
```

Import data file and create arrays that are in the order of the MBTA stops (using the names of the stops from the .csv)

```python
tdatafile=open('LateNight_thru_7June2014.csv')
tcsv=pd.read_csv(tdatafile)
```

```python
redlinestations=['Alewife','Davis Square','Porter Square','Harvard','Central Square','Kendall Square','Charles MGH','Park Street','Downtown Crossing',
                 'South Station','Broadway','Andrew Square','JFK/U Mass','North Quincy','Wollaston','Quincy Center','Quincy Adams','Braintree',
                 'Savin Hill','Fields Corner','Shawmut','Ashmont','Mattapan Line']

bluelinestations=['Wonderland','Revere Beach','Beachmont','Suffolk Downs','Orient Heights','Wood Island','Airport','Maverick','Aquarium','State Street','Government Center','Bowdoin']

greenlinestations=['Lechmere','Science Park','North Station','Haymarket','Government Center','Hynes','Park Street','Boylston','Arlington',
                   'Copley Square','Wellington ','Hynes','Kenmore Square','Prudential','Symphony','Green Line B','Green Line C','Green Line D','Green Line E']

orangelinestations=['Oak Grove','Malden Center ','Wellington ','Sullivan Square','Community College','North Station','Haymarket','State Street',
                    'Downtown Crossing','Chinatown','Tufts Medical Center','Back Bay','Mass Ave','Ruggles','Roxbury Crossing','Jackson Square','Stony Brook','Green Street','Forest Hills']
```

Trim down the csv so that we can focus only on the train lines being used in this analysis.

```python
trainsonly=tcsv[tcsv['line'].isin(['Red','Blue','Green','Orange'])]
```

Perform a "Group By" on the train data so that categorize the total number of transactions by hour, by station, by line (in the order in which we're performing the group by, so line first, then stations, then by hour)

```python
grouped=trainsonly.groupby(['line','routestation','trxhour'],as_index=False)['transactions'].sum()
```

Create data frames that are isolated to each of the lines so that we can visualize one line at a time using the pcolor charts in matplotlib

```python
redline=grouped[grouped['routestation'].isin(redlinestations)]
blueline=grouped[grouped['routestation'].isin(bluelinestations)]
orangeline=grouped[grouped['routestation'].isin(orangelinestations)]
greenline=grouped[grouped['routestation'].isin(greenlinestations)]
```

Pivot the data so that each pair of station and hour has a value (number of transactions, giving us a nice 2D matrix)

```python
rf=redline.pivot(index='routestation', columns='trxhour', values='transactions')
bf=blueline.pivot(index='routestation', columns='trxhour', values='transactions')
gf=greenline.pivot(index='routestation', columns='trxhour', values='transactions')
of=orangeline.pivot(index='routestation', columns='trxhour', values='transactions')
```

Because pivoting moved everything to alphabetical order, we're going to "reindex" the row order against the list of stations we made earlier. Also, convert those dataframes into Numpy matrices.

```python
rdf=rf.reindex(index=redlinestations)
bdf=bf.reindex(index=bluelinestations)
gdf=gf.reindex(index=greenlinestations)
odf=of.reindex(index=orangelinestations)
```

1. Create an iterable array of all of the dataframes (so we don't have to recreate the plots over and over again) as well as add an array for each of the titles. Also create variables that will be used for the row labels and column labels of the visualization.
2. Create array of colors (colors)
3. Render the visualization

```python
dataframes=[rdf,bdf,gdf,odf]
titles=['Red Line','Blue Line','Green Line','Orange Line']
colors=[plt.cm.Reds,plt.cm.Blues,plt.cm.Greens,plt.cm.Oranges]
row_labels=[]
column_labels=[]
data=[]
# Create row and column labels for each visualization that represent the labels, as well as convert the dataframes into Numpy Matrices

for frame in dataframes:
    row_labels.append(frame.index)
    column_labels.append(['10PM','11PM','12AM','1AM','2AM'])
    data.append(frame.as_matrix())



# Create a set of subplots with the number of dataframes as the subplots.
fig, ax = plt.subplots(nrows=len(dataframes),sharex=False)

for num in range(0,len(data)):
       ax[num].pcolor(data[num],cmap=colors[num],label='Banana')
fig.frameon=False
fig.set_figheight(30)
fig.set_figwidth(15)

# put the major ticks at the middle of each cell
for var in range(0,len(ax)):
    
    ax[var].set_xticks(np.arange(dataframes[var].shape[1])+.5, minor=False)
    ax[var].set_yticks(np.arange(dataframes[var].shape[0])+.5, minor=False)
    ax[var].set_title(titles[var],size=20,fontname='Adobe Garamond Pro')
    ax[var].set_adjustable('box-forced')


    ax[var].set_xticklabels(column_labels[var], minor=False,size=15,fontname='Adobe Garamond Pro')
    ax[var].set_yticklabels(row_labels[var], minor=False,size=20,fontname='Adobe Garamond Pro',weight='light')
    #ax.xaxis.set_ticks([])
    #ax.yaxis.set_ticks([])
    for spine in ax[var].spines.itervalues():
       spine.set_visible(False)
    for t in ax[var].xaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
    for t in ax[var].yaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
plt.legend()
plt.tight_layout()
plt.autoscale()

#fig.subplots_adjust(hspace=.1)
plt.savefig('LateNightT-Visualized.png')
```
